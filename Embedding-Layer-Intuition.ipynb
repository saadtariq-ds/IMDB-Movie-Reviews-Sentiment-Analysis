## Import Libraries

```python
import numpy as np
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras import optimizers, losses
```

## Data

```python
sentences = [
    'the glass of milk',
    'the glass of juice',
    'the cup of tea',
    'I am a boy',
    'I am a good developer',
    'understand the meaning of words',
    'your videos are good',
]
```

```python
sentences
```

## Define the Vocabulary Size

```python
vocabulary_size = 10000
```

```python
one_hot_representation = [one_hot(words, vocabulary_size) for words in sentences]
one_hot_representation
```

## Word Embedding Representation

```python
sentence_length = 8
```

```python
embedded_documents = pad_sequences(
    one_hot_representation,
    padding='pre',
    # padding='post',
    maxlen=sentence_length
)

embedded_documents
```

## Feature Representation and Model Training

```python
dimensions = 10
```

```python
optimizer = optimizers.Adam(learning_rate=0.001)
loss = losses.MeanSquaredError()
```

```python
model = Sequential()

model.add(Embedding(input_dim=vocabulary_size, output_dim=dimensions, input_shape=(sentence_length,)))

model.compile(
    optimizer=optimizer,
    loss=loss
)
```

```python
model.summary()
```

```python
model.predict(embedded_documents)
```
